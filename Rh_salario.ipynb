```python
# Importar as libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('salario_rh.csv')

df.head()
```

```python
# Renomenado as colunas
df.rename( columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'
}, inplace=True )
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Grafico Nulos
sns.heatmap( df.isnull(), cbar=False );
```

```python
df.describe()
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise da Renda')
sns.kdeplot( df['Renda'], color='green' );
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise da Xp')
sns.kdeplot( df['Xp'], color='blue' );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.boxplot( df['Renda'] );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.boxplot( df['Xp'] );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.scatterplot( data=df, x='Renda', y='Xp');
```

```python
plt.figure( figsize=(10, 5) )
sns.regplot( data=df, x='Renda', y='Xp');
```

```python
# Correlação
Correlacao = df.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```

```python
# Correlação
Correlacao = df.corr('spearman')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
